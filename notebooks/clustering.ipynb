```python
import numpy as np
from scipy.spatial.distance import cdist
```

```python
def leader_clustering(X, threshold: float):

    centroid_indices = []
    labels = np.full(shape=X.shape[0], fill_value=-1, dtype=int)
    visited = np.zeros(shape=X.shape[0], dtype=bool)
    current_label = 0

    for i in range(X.shape[0]):
        if visited[i]:
            continue

        centroid_indices.append(i)
        labels[i] = current_label
        visited[i] = True

        for j in range(i + 1, X.shape[0]):
            if not visited[j]:
                distance = np.linalg.norm(X[i] - X[j])
                if distance < threshold:
                    labels[j] = current_label
                    visited[j] = True

        current_label += 1
```

```python
points = np.array([
    [0, 1, 0],
    [1, 2, 0],
    [5, 10, 0],
    [1, 11, 4],
])

centroids = np.array([
    [0, 2, 1],
    [0, 8, 0],
])

distances = cdist(points, centroids)
```
